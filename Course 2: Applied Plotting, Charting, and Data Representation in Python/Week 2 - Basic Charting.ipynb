```python
# import the artist class from matplotlib
from matplotlib.artist import Artist

def rec_gc(art, depth=0):
    if isinstance(art, Artist):
        # increase the depth for pretty printing
        print("  " * depth + str(art))
        for child in art.get_children():
            rec_gc(child, depth+2)

# Call this function on the legend artist to see what the legend is made up of
rec_gc(plt.legend())
```

## Line Plots

```python
linear_data = np.array([1, 2, 3, 4, 5, 6, 7, 8])
quadratic_data = linear_data ** 2

plt.figure()
# will use the index as x-values
plt.plot(linear_data, '-o', quadtratic_data, '-o')
```

```python
plt.plot([22, 44, 55], '--r')
```

```python
plt.xlabel('Some data')
plt.ylabel('Some more data')
plt.title('Title')
plt.legend(['Baseline', 'Competition', 'Us'])
```

```python
plt.gca().fill_between(range(len(linear_data)), linear_data, quadratic_data, facecolor='blue', alpha=0.25)
```

```python
# new figure with date time x-axis
plt.figure()

observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')
plt.plot(observation_dates, linear_data, '-o', observation_dates, quadratic_data, '-o')
```

```python
# date time handling is a pain point in data science. Numpy, matplotlib, and pandas use different types
import pandas as pd

plt.figure()
observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')
observation_dates = list(map(pd.to_datetime, observation_dates))  # matplotlib cant use a generator, so convert to list
plt.plot(observation_dates, linear_data, '-o',
        observation_dates, quadratic_data, '-o')
```

```python
x = plt.gca().xaxis

for item in x.get_ticklabels():
    item.set_rotation(45)
```

```python
plt.subplots_adjust(bottom=0.25)
```

```python
ax = plt.gca()
ax.set_xlabel('Date', color='white')
ax.set_ylabel('Units', color='white')
ax.set_title('Quadtratic vs Linear Performace', color='white')
```

## Bar Charts

```python
plt.figure()
xvals = range(len(linear_data))
plt.bar(xvals, linear_data, width=0.3)
```

```python
new_xvals = []
for item in xvals:
    new_xvals.append(item + 0.3)
    
plt.bar(new_xvals, quadratic_data, width=0.3, color='red')
```

```python
# errors
from random import randint
linear_err = [randint(1, 5) for x in range(len(linear_data))]

plt.bar(xvals, linear_data, width=0.3, yerr=linear_err)
```

```python
# stacked bar charts
plt.figure()
xvals = range(len(linear_data))
plt.bar(xvals, linear_data, width=0.3, color='b')
# set bottom = same y data above
plt.bar(xvals, quadtratic_data, width=0.3, bottom=linear_data, color='r')
```

```python
# pivot to a horizontal
plt.figure()
xvals = range(len(linear_data))
plt.barh(xvals, linear_data, height=0.3, color='b')
plt.barh(xvals, quadtratic_data, height=0.3, left=linear_data, color='r')
```

## Dejunkifying a Plot

### Exercises: 

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()

languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]

# change all bars to lighter grey
bars = plt.bar(pos, popularity, align='center', linewidth=0, color='lightslategrey')
plt.xticks(pos, languages, color='lightgrey', alpha=0.8)  # soften ticks
# Remove
# plt.ylabel('% Popularity', color='lightgrey', alpha=0.8)  # soften labels
plt.title('Top 5 Languages for Math & Data \nby % popularity on Stack Overflow', alpha=0.8, color='lightgrey')  # soften title

# 1) Remove all ticks and tick labels on y-axis
plt.tick_params(
    axis='both',       
    which='both',      
    bottom='off',      
    left='off',
    labelleft='off')

# 2) Remove frame 
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.set_facecolor('none')


# Instructor solution
for spine in plt.gca().spines.values():
    spine.set_visible(False)

# 3) Change the bar colors to be less bright blue, make the python a contrasting color, soften all labels by turning grey
# Other solutions are in higher lines
bars[0].set_color('darkorange')

# 4) Directly label y values
for bar in bars:
    plt.gca().text(bar.get_x() + bar.get_width()/2,        # x-coor
                   bar.get_height() - 4,                   # y-coor
                   str(int(bar.get_height())) + '%',       # text
                   ha='center', color='white', fontsize=12)

plt.show()
```
