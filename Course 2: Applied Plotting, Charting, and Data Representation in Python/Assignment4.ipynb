#### What are the major religions in the United States from 1945 - 2010?

```python
usa_top = usa.filter(regex='_all').mean().sort_values(ascending=False)
usa_top = usa_top.drop('religion_all')
usa_top.head(5)
```

### USA

```python
christianity = usa.christianity_percent
noreligion = usa.noreligion_percent
judaism = usa.judaism_percent
buddhism = usa.buddhism_percent
islam = usa.islam_percent
```

```python
plt.figure(figsize=(14,8))

plt.plot(christianity, 'red',
        noreligion, 'blue',
        judaism, 'green', 
        buddhism, 'purple',
        islam, 'yellow')

plt.title('United States Religion Adherence %', color='lightgray', size=15, alpha=0.8)
plt.tick_params(axis='both', color='none', labelcolor='lightgray')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.set_facecolor('none')
ax.set_ylim(0, 1)
plt.setp(ax.xaxis.get_majorticklabels(), ha='right')
plt.grid(axis='y', alpha=0.1)

x = plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(45)

plt.show()
```

#### Puting both plots into one figure and clean up code

```python
# Series
christianity = usa.christianity_percent
noreligion = usa.noreligion_percent
judaism = usa.judaism_percent
buddhism = usa.buddhism_percent
islam = usa.islam_percent
usa_data = [christianity, noreligion, judaism, buddhism, islam]

g_christianity = global_.christianity_percent
g_noreligion = global_.noreligion_percent
g_judaism = global_.judaism_percent
g_buddhism = global_.buddhism_percent
g_islam = global_.islam_percent
global_data = [g_christianity, g_noreligion, g_judaism, g_buddhism, g_islam]

all_data = [usa_data, global_data]

# Figure
fig, (ax_usa, ax_global) = plt.subplots(1, 2, sharex=True, sharey=True, figsize=(20, 8), facecolor='#292929')
fig.subplots_adjust(top=0.75)
plt.subplots_adjust(wspace=0.1)

# Plotting variables
colors = ['red', 'blue', 'green', 'purple', 'yellow']
titles = ['United States', 'Global']
labels = ['Christianity', 'No Religion', 'Judaism', 'Buddhism', 'Islam']
axes = [ax_usa, ax_global]

# Figure common title
fig.suptitle("How has the percentage of adherents to the most common \n \
religions in the United States changed from 1945-2010? \n \
(Compare to the rest of the world)", 
             fontsize=20,
            color='lightgray')

# Plot data
# Loop through both axes
for i in range(0, len(axes)):
    ax = axes[i]
    
    # Loop through all series and plot lines on their respective axes
    for j in range(0, len(usa_data)):
        axes[i].plot(all_data[i][j], colors[j])
    
    # Adjustments
    ax.set_title(titles[i], color='lightgray', size=18, alpha=0.8, va='bottom')
    ax.tick_params(axis='both', color='none', labelcolor='lightgray')
    ax.set_facecolor('none')
    ax.set_ylim(0, 1)
    ax.grid(axis='y', alpha=0.1)
    
    # Set all spines to invisible
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)

    # Align tick labels to right
    for tick in ax.xaxis.get_majorticklabels():
        tick.set_horizontalalignment("right")
    
    # Rotate x-axis labels
    xlabels = ax.xaxis.get_ticklabels()
    for item in xlabels:
        item.set_rotation(45)
                
# Add legend
legend = plt.legend(labels=labels, 
                    loc=8, 
                    ncol=5, 
                    fontsize=16, 
                    facecolor='#434343', 
                    bbox_to_anchor=(-0.1,-0.3), 
                    frameon=False)

# Set font color
plt.setp(legend.get_texts(), color='lightgray')

# Increase width of legend markers
for legobj in legend.legendHandles:
    legobj.set_linewidth(5)
```
