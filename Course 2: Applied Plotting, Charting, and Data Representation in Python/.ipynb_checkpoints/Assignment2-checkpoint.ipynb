```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
import mplleaflet
import pandas as pd

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'21f22d18c15db96f2a6ce96fdf640ced045b79a0fcb2ceacf61668e3')
```

```python
%matplotlib inline
import numpy as np
import matplotlib.dates as mdates
import matplotlib.ticker as ticker

# Load data
df = pd.read_csv('data/C2A2_data/BinnedCsvs_d400/21f22d18c15db96f2a6ce96fdf640ced045b79a0fcb2ceacf61668e3.csv')
df = df.sort_values(['Date', 'ID'])
```

```python
# Re-arrange Data_Value to separate columns
TMAX = df[df['Element'] == 'TMAX']
TMIN = df[df['Element'] == 'TMIN']
merge = pd.merge(TMAX, TMIN, how='inner', on=['ID', 'Date'])
merge = merge.drop(['ID', 'Element_x', 'Element_y'], axis=1)
merge.columns = ['DATE', 'TMAX', 'TMIN']
merge.TMAX /= 10
merge.TMIN /= 10

# Remove leap days (02-29)
merge['YEAR'] = merge['DATE'].str[:4]
merge['MO-DY'] = merge['DATE'].str[5:]
merge = merge[merge['MO-DY'] != '02-29']

# Make DataFrames for 2005-2014 and 2015 data
five_fourteen = merge[((merge['YEAR']) >= '2005') & ((merge['YEAR']) <= '2014')]
fifteen = merge[merge['YEAR'] == '2015']

# Aggregate Max of TMAX and Min of TMIN by day of year
five_fourteen = five_fourteen.groupby('MO-DY').agg({'TMAX': 'max', 'TMIN': 'min'})
fifteen = fifteen.groupby('MO-DY').agg({'TMAX': 'max', 'TMIN': 'min'})

# Filter 2015 data for TMAX and TMIN means outside of 2005-2014 means
fifteen['TMAX'] = fifteen[fifteen['TMAX'] > five_fourteen['TMAX']]['TMAX']
fifteen['TMIN'] = fifteen[fifteen['TMIN'] < five_fourteen['TMIN']]['TMIN']

# Plot 2005-2014 data
plt.figure(figsize=(10, 6), dpi=80, facecolor='#393939')
plt.plot(five_fourteen['TMAX'], '-', lw=1.5, color='indianred')
plt.plot(five_fourteen['TMIN'], '-', lw=1.5, color='C')
plt.gca().fill_between(range(len(five_fourteen)), five_fourteen['TMIN'], five_fourteen['TMAX'], facecolor='black', alpha=0.2)

# Overlay 2015 data
plt.plot(fifteen['TMAX'], '^', color='firebrick')
plt.plot(fifteen['TMIN'], 'v', color='darkslategray')

# Axes
ax = plt.gca()
ax.grid(visible=False)
plt.
plt.ylabel('Celsius', fontsize=14, color='white')
plt.title('Los Angeles Temperatures 2015', fontsize=16, color='white')
plt.xticks(color='white')
# yvals = [-50, -40, -30, -20, -10, 0, 10, 20, 30, 40, 50]
plt.yticks(color='white')

day_intervals = [0, 30, 60, 90, 120, 150, 180, 210, 240, 270, 300, 330]
adjusted = [i + 15 for i in a]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(adjusted, months)

# Legend
high_count = fifteen['TMAX'].count()
low_count = fifteen['TMIN'].count()
plt.legend(['2005-2014 High', '2005-2014 Low', 'Record High ({})'.format(high_count), 'Record Low ({})'.format(low_count)], 
           frameon=True, 
           fontsize='large', 
           fancybox=True, 
           loc='best')

plt.savefig('temp.png')
plt.show()
```
